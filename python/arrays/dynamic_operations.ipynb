### Dynamic Arrays Operations


```python
from core import Array

class DynamicArray():
    def __init__(self, initial_capacity = 1, typecode = 'I'):
        self._array = Array(initial_capacity, typecode)
        self._capacity = initial_capacity
        self._size = 0
        self._typecode = typecode

    def __len__(self) -> int:
        return self._size
    
    def __repr__(self) -> str:
        return repr(self._array._array[:self._size])

    def _double_size(self):
        old_array = self._array
        self._array = Array(self._capacity * 2, self._typecode)
        self._capacity *= 2
        for i in range(self._size):
            self._array[i] = old_array[i]
    
    def _half_size(self):
        new_capacity = max(1, self._capacity // 2)
        if self._size > new_capacity:
            return  # Don't shrink if size won't fit

        old_array = self._array
        self._array = Array(new_capacity, self._typecode)
        self._capacity = new_capacity
        for i in range(self._size):
            self._array[i] = old_array[i]
    
    def insert(self, value):
        if self._size >= self._capacity:
            self._double_size()
        self._array[self._size] = value
        self._size += 1

    def find(self, target):
        if self._size == 0:
            raise ValueError(f'Empty Array')
        for i in range(self._size):
            if self._array[i] == target:
                return i
        return None

    def delete_by_value(self, target):
        index = self.find(target)
        print(f"index to delete: {index} and its Number is: {target}")
        if index is None:
            raise ValueError(f'{target} not found in array')
        for i in range(index, self._size - 1):
            self._array[i] = self._array[i + 1]
        self._size -= 1
        
        print(f'capacity {self._capacity}')
        print(f'self size: {self._size}')
        
        if self._capacity > 1 and self._size <= self._capacity:
           self._half_size()



# Implement the delete-by-index method and make sure that the amortized running time and extra space for n deletions are both O(n). Hint: Suppose we donâ€™t need to preserve the insertion order of the elements.

    def delete_by_index(self, index):
        if self._size == 0:
            raise ValueError('Empty Array')
        if index < 0 or index >= self._size:
            raise IndexError(f'Index:{index} is out of range')

        self._array[index] = self._array[self._size - 1]
        self._size -= 1
        if self._capacity > 1 and self._size <= self._capacity:
            self._half_size()

        
        
        

a = DynamicArray()
a.insert(1)
a.insert(2)
a.insert(3)
a.insert(4)
a.insert(5)
print(a._array)
a.insert(6)
print(a._array)
print(a.find(3))
a.delete_by_value(2)
print(a)
a.delete_by_index(0)
print(a)




```
