```python
from core import Array

class myUnsortedArray:
    def __init__(self, max_size: int, typecode: str ='l'):
        self._array = Array(max_size, typecode)   # Actual Storage space (fixed size) (shelf/container)
        self._max_size = max_size
        self._size = 0            #Items currently in the shelf

    def __len__(self) -> int:
        return self._size 
    
    def __repr__(self) -> str:
        '''
            Returns the string representation of the array
            Parameters: None
        '''
        return f'UnsortedArray({repr(self._array._array[:self._size])})'
    
    def __getitem__(self, index):
        if index < 0 or index >= self._size:
            raise IndexError(f'Index: {index} out of range.')
        return self._array[index]

    def max_size(self) -> int:
        return self._max_size
    
    def insert(self, value) -> None:
        if self._size >= len(self._array): 
            raise ValueError(f'Full Array')
        else:
            self._array[self._size] = value
            self._size += 1
    
 
    
    def delete(self, index) -> None:
        if self._size == 0:
            raise ValueError(f'Empty Array')
        elif index < 0 and index >= self._size:
            raise ValueError(f'Out of range')
        else:
            self._array[index] = self._array[self._size - 1]
            self._size -= 1


    def search(self, target) -> int:
        if self._size == 0:
            raise ValueError(f'Empty Array')
        else:
            for i in range(self._size):
                if self._array[i] == target:
                    return i
            return -1
        
    def traverse(self, callback):
        if self._size == 0:
            raise ValueError(f'Empty Array')
        for i in range(self._size):
            callback(self._array[i])


    def max_in_array(self) -> int:
        if self._size == 0:
            raise ValueError(f'Empty Array')
        max_num = self._array[0]
        for i in range(self._size):
            if max_num < self._array[i]:
                max_num = self._array[i]
        return max_num


    def min_in_array(self) -> int:
        if self._size == 0:
            raise ValueError(f'Empty Array')
        min_index = 0
        for i in range(self._size):
            if self._array[min_index] > self._array[i]:
                min_index = i
        return self._array[min_index]
        

    def sum(self) -> int:
        if self._size == 0:
            raise ValueError(f'Empty Array')
        total = 0
        for index in range(self._size):
           total += self._array[index]
        return total


a = myUnsortedArray(5)
a.insert(3)
a.insert(43)
a.insert(8)
a.insert(25)
print(f'Total Count: {len(a)}')
a.traverse(print)
print(f'Sum: {a.sum()}')
print(f'All: {a}')
print(f'Max in array: {a.max_in_array()}')
print(f'Min in array: {a.min_in_array()}')
a.delete(2)
print(f'Index of 43: {a.search(43)}')
```
