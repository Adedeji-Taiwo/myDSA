### Day 8: Search and Update Nodes

- How to search for a value in the list
- How to update a node’s value
- How to handle cases where the value isn’t found



```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None # Start of list (empty)

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")


    def search(self, target):
        if self.head is None:
            print(f'Empty List')
            return
        
        
        if self.head.data == target:
            print(f'Target: {target} found in first node')
            return
        
        current = self.head
        while current.next is not None:
            
            if current.next.data == target:
                print(f'Target: {target} found')
                return
            current = current.next
            
        print(f'Target: {target} not found')



    def search_v2(self, target):
        current = self.head
        
        while current:
            if current.data == target:
                print(f'Target: {target} found')
                return
            current = current.next
        print(f'Target: {target} not found')


    
    def update(self, target, replacement):
        if self.head == None:
            print(f'Empty list')
            return

        if self.head.data == target:
            self.head.data = replacement
            print(f'First node" data: {target} replace with {replacement}')       
            return

        current = self.head
        while current.next is not None:
            if current.next.data == target:
                current.next.data = replacement
                print(f'Target: {target} replaced with {replacement}') 
            current = current.next
        print(f'Target: {target} not found. Hence can"t be replaced')



    def update_v2(self, old_value, new_value):
        current = self.head

        while current:
            if current.data == old_value:
                current.data = new_value
                print(f'Value replaced successfulyy')
                return
            current = current.next
        print(f'Value to be replaced not found in list')


    def double_values(self, callback):
        current = self.head
        box = []

        while current:
            if current.data is not None:
                box.append(int(current.data) * 2)
            current = current.next
        callback(box)


# Manually create nodes
node1 = Node(5)
node2 = Node(8)
node3 = Node(10)

# Link Nodes
node1.next = node2
node2.next = node3

a = LinkedList()
a.head = node1

a.display()
a.search(8)
a.update(5, 6)
a.update_v2(15, 16)
a.display()
a.double_values(print)


            
```
